## Calculate PH for each conformation

```python
#export
def createDiagramSetPH(p:atompartitioning.AtomPartition, strucname:str="protein"):
    '''
    Computes PH diagram set using Ripser by appending them together.
    '''
    # create a new graph object for each partition
    import matplotlib.pyplot as plt

    fig, axs = plt.subplots(1, 4)
    fig.suptitle(f'Persistence diagrams for structure {strucname}')
    
    count = 1
    for i in range(len(axs)):
        structuralPH.printPHDiagram(p.byaatype[i],str(count))
        count = count+1
        axs[i].plot()
    
    fig.savefig(f"{strucname}.png")
    
    return
```
